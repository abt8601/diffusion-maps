```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from diffusion_maps import diffusion_maps
```

# Generate Helix Data

```python
n_samples = 1000
t = np.linspace(0, 8 * np.pi, n_samples)
x = np.cos(t)
y = np.sin(t)
z = t / (4 * np.pi) - 1
helix = np.column_stack((x, y, z))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=np.linspace(0, 1, 1000), cmap='viridis')
plt.show()
```

# Compute Diffusion Maps

```python
mapped_helix = diffusion_maps(helix, n_components=2, kernel='gaussian', sigma=0.1, diffusion_time=1)
```

```python
plt.scatter(mapped_helix[:, 0], mapped_helix[:, 1], c=np.linspace(0, 1, 1000), cmap='viridis')
plt.axis('square')
plt.show()
```
